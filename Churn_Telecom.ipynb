**I.INTRODUCTION**

Présentation du Projet

* L’environnement des entreprises est marqué par de profondes mutations. En effet, les entreprises doivent faire face à deux principaux défis : d’une part, l’exigence de réactivité face à une mondialisation des marchés, à une intensité concurrentielle croissante et au pouvoir grandissant des clients, et d’autre part, un fort développement des technologies de l’information et de la communication favorisant les échanges tant à l’intérieur d’une organisation qu’à l’extérieur.
* Par rapport à ça Tunisie Telecom a lancer un projet appelé "Ichmilni" qui a pour but d'appliquer un programme nommé "taghtia" pour réunir tout les tunisiens sur l'ensemble du territoire international.
* Il vise en effet à démocratiser l'accés à l'information , à palier à "l'itterisme électronique" et à donc transmettre à chaque individu les compétences numériques les plus élementaires qui seront un levier de son inclusion sociale et économique.

Qu'est-ce que le taux de désabonnement ?
* Le taux de désabonnement des clients est défini comme le moment où les clients ou les abonnés cessent de faire affaire avec une entreprise ou un service.

* Les clients de l'industrie des télécommunications peuvent choisir parmi une variété de fournisseurs de services et passer activement de l'un à l'autre. Le secteur des télécommunications a un taux de roulement annuel de 15 à 25 % sur ce marché hautement concurrentiel.

* La fidélisation personnalisée des clients est difficile car la plupart des entreprises ont un grand nombre de clients et ne peuvent pas se permettre de consacrer beaucoup de temps à chacun d'eux. Les coûts seraient trop importants, dépassant les revenus supplémentaires. Cependant, si une entreprise pouvait prévoir à l'avance quels clients sont susceptibles de partir, elle pourrait concentrer ses efforts de fidélisation de la clientèle uniquement sur ces clients «à haut risque». Le but ultime est d'étendre sa zone de couverture et de fidéliser davantage ses clients. Le cœur du succès sur ce marché réside dans le client lui-même.

* Le taux de désabonnement des clients est une mesure essentielle car il est beaucoup moins coûteux de fidéliser les clients existants que d'en acquérir de nouveaux.


* Pour réduire l'attrition des clients, les entreprises de télécommunications doivent prévoir quels clients sont à haut risque d'attrition.

* Pour détecter les premiers signes d'attrition potentielle, il faut d'abord développer une vision holistique des clients et de leurs interactions sur de nombreux canaux, y compris les visites en magasin/succursales, les historiques d'achat de produits, les appels au service client, les transactions sur le Web et les interactions avec les médias sociaux, pour citer quelques-uns.

* Par conséquent, en s'attaquant au désabonnement, ces entreprises peuvent non seulement préserver leur position sur le marché, mais aussi croître et prospérer. Plus ils ont de clients dans leur réseau, plus le coût d'initiation est bas et plus le profit est important. 
* En conclusion, l'objectif principal de l'entreprise pour réussir est de réduire l'attrition des clients et de mettre en œuvre une stratégie de fidélisation efficace.

# IV.1.Logistic Regression

```python
lr_model = LogisticRegression()
lr_model.fit(X_train,y_train)
accuracy_lr = lr_model.score(X_test,y_test)
print("Logistic Regression accuracy is :",accuracy_lr)
```

# IV.6.VotingClassifier

* Prédisons maintenant le modèle final basé sur la plus grande majorité des votes et vérifions son score.

```python
from sklearn.ensemble import VotingClassifier
clf1 = GradientBoostingClassifier()
clf2 = LogisticRegression()
clf3 = AdaBoostClassifier()
clf4= RandomForestClassifier()
eclf1 = VotingClassifier(estimators=[('gbc', clf1), ('lr', clf2), ('abc', clf3),('rfc',clf4)], voting='soft')
eclf1.fit(X_train, y_train)
predictions = eclf1.predict(X_test)
print("Final Accuracy Score ")
print(accuracy_score(y_test, predictions))
```
